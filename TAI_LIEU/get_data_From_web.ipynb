```python
from bs4 import BeautifulSoup
import requests
import re
```

# Get data from website with selenium

```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.keys import Keys
import random
from selenium.webdriver.support.ui import WebDriverWait

driver = webdriver.Chrome()
driver.get("https://www.vietnamworks.com/data-analyst-kv")

sleep(random.randint(5,10))

btn = driver.find_element(by="xpath", value='/html/body/div/div[6]/div[2]/div[1]/main/div/div[1]/div[1]/div[4]/div/div[51]/ul/li[1]/a')
btn.click()

block_list = driver.find_element(by='xpath', value="//div[@class='block-job-list']")
list_job = block_list.find_elements(by='xpath', value="//a[@href]")
```

# Conver data raw to csv

### Clean data

```python
data = []
for inx, job in enumerate(list_job):
    url = job.get_attribute('href')
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        if ("job-title" in response.text) and ("name" in response.text) and ("company-location" in response.text) and ("salary" in response.text) and ("expiry gray-light" in response.text) and ("benefit row" in response.text) and ("job-description mobile-box" in response.text) and ("requirements" in response.text):
            job_title = soup.find(class_ = "job-title")
            company = soup.find(class_ = "name")
            location = soup.find(class_ = "company-location")
            salary = soup.find(class_ = "salary")
            deadline = soup.find(class_ = "expiry gray-light")
            benefit = soup.find_all(class_ = "benefit row")
            str_benefit = "\n- ".join(line.text.strip() for line in benefit)
            desc = soup.find(class_ = "description")
            requirement = soup.find(class_ = "requirements")

            # data.append([job_title,company,location,salary,deadline,str_benefit,desc,requirement])
            data.append([job_title.text.strip() ,company.text.strip(),location.text.strip(),salary.text.strip(),deadline.text.strip(),str_benefit,desc.text.strip(),requirement.text.strip()])
    else: 
        print("Error", response.status_code)
        print(inx)

```

### Convert to csv

```python
import pandas as pd
df = pd.DataFrame(data, columns=['Tên Job', "Tên công ty", "Địa điểm", "Lương", "Deadline", "Phúc lợi", "Mô tả công việc", "Yêu cầu công việc"])
df = df.drop_duplicates(subset= ["Tên công ty","Tên Job"])
```
