```python
scraper = cloudscraper.create_scraper(delay=5, browser="chrome")
player_data = []
for offset in range(0,20000,60):
    url = f"https://sofifa.com/players?offset={offset}" 
    content = scraper.get(url)
    if content.status_code == 200:
        soup = BeautifulSoup(content.text, 'html.parser')
        parsed_data1 = soup.find_all("a",{'aria-label': True})
        for data in parsed_data1:
            aria_label = data['href']
            player_data.append("https://sofifa.com/" + aria_label)
    else:
        print("Error",content.status_code,"offset:",offset)
        

player_data
```

```python
import pandas as pd
df = pd.DataFrame({'Urls':player_data})
```

```python
df
```

```python
df = df[df['Urls'].str.contains('/player/')]
df = df.reset_index(drop=True)
df
```

```python
df.to_csv("D:/Jupyter/DATA/crawl_url.csv", encoding='utf-8-sig')
```
